```python
import requests
from datetime import datetime, timezone

def fetch_latest_umasep100():
    ts = datetime.now(timezone.utc).strftime("%Y-%m-%dT%H:%M:%SZ")
    url = "https://iswa.gsfc.nasa.gov/IswaSystemWebApp/FileRedirect"
    params = {
        "dataID": 1210,  # UMASEP-100
        "search": "lessThanOrEqualTo",
        "timestamp": ts
    }
    r = requests.get(url, params=params, timeout=20, allow_redirects=True)
    r.raise_for_status()
    return r.json(), r.url  # (JSON내용, 실제로 리다이렉트된 파일 URL)

data, resolved_url = fetch_latest_umasep100()
```
