```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import pickle
from PIL import Image
import os

year = 2002

hproton = pd.read_csv(
    f"{year}_DPD.txt",
    delim_whitespace=True,
    comment="#",
    header=None,
    names=[
        "year", "month", "day",
        "p_gt_1MeV", "p_gt_10MeV", "p_gt_100MeV",
        "e_gt_0p6MeV", "e_gt_2MeV",
        "neutron_pct"
    ]
)

hproton_100MeV = hproton["p_gt_100MeV"].copy().astype(float)
```

```python
pickle.dump(hproton_100MeV.values[1:], open(f"hproton_{year}.pkl", "wb"))
```
